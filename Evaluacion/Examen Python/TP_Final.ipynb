## TEMÁTICA: ANÁLISIS ESTRUCTURAL Y GESTIÓN DE OBRA CIVIL EN VILLA MARÍA

### Introducción
La municipalidad de Villa María comenzó un proyecto de renovación de la infraestructura urbana que incluye:
- Construcción de un nuevo puente sobre el río Ctalamochita, en barrio Palermo.
- Repavimentación de calles
- Análisis de desagües pluviales

El equipo de ingenieros de la dirección de obras públicas de la municipalidad debe desarrollar un sistema informático para analizar y gestionar estos proyectos.
