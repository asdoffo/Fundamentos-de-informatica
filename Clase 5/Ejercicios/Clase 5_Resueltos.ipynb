**a.** Hacé un código que lea un csv como el mencionado anteriormente y devuelva una **lista de diccionarios** conteniendo los datos de cada fila, donde cada encabezado es la clave del valor correspondiente. Por ejemplo, el primer elemento de la lista debería ser `{'ticker': 'LOMA', 'fecha': '2021-04-26', 'cantidad': '28', 'precio_compra': '184.3', 'precio_actual': '507'}`.

*Ayuda: la función zip puede facilitar la creación de los diccionarios.*

```python
import csv
# antes de ejecurar esto me aseguro que subi en la ruta correcta el archivo
# Luego, abro el archivo dando la ruta completa
with open("/content/drive/MyDrive/Colab Notebooks/Data/portafolio_juan.csv") as f:         # Trabajo con el archivo portafolio_juan.csv que está subido a drive (en este caso) dentro de with 
  filas = csv.reader(f)                                               # Leo e interpreto todo el contenido del archivo
  encabezados = next(filas)                                           # Guardo y salteo la primera fila
  acciones = []
  for fila in filas:                                                  # Recorro todo el contenido del archivo: cada fila es una lista de características de una acción
    acciones.append(dict(zip(encabezados, fila)))             # Adjunto a la lista acciones un diccionario generado con un zip de los encabezados y las filas
  print(acciones)                                                     
```

**b.** A partir de la lista de diccionarios que creaste en el inciso anterior, calculá el valor actual del portafolio de acciones (es decir, la suma del precio actual de las acciones mutiplicado por la cantidad de acciones que posee) y el costo de compra del mismo (la suma del precio de compra de las acciones mutiplicado por la cantidad de acciones que compró ese día). ¿Juan ganó o perdió dinero?

```python
valor_actual_acciones = 0
costo_total_acciones = 0                                                        
for accion in acciones:                                                     # Para cada acción:
  valor_actual_acciones += float(accion["precio_actual"]) * float(accion["cantidad"])                 # Calculo el valor actual total (si vendiera todas las acciones "ahora")
  costo_total_acciones += float(accion["precio_compra"]) * float(accion["cantidad"])                  # Calculo el costo total de compra con el que compró las acciones antes

ganancia_acciones = valor_actual_acciones - costo_total_acciones                                                        # Sumo la diferencia (la ganancia o la pérdida) al portafolio

print(f"La ganancia neta que obtuvo Juan con su portafolio es de {round(ganancia_acciones,2)}")
if ganancia_acciones > 0:
  print("Juan ganó dinero")
else:
  print("Juan no ganó dinero")
```
