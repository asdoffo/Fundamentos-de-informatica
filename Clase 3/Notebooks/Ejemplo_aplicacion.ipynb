Con el contenido hasta la clase 3, podemos realizar un pequeño script de ejemplo:

### "Cifrado secreto y sescifrado con Python"

```python

# SCRIPT: CIFRADO Y DESCIFRADO BÁSICO CON PYTHON

# Usamos una lista (el abecedario) para trabajar con índices
ABECEDARIO = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
# Se utiliza un diccionario para mapear letras a sus posiciones
CLAVE_CESAR = 3  # Clave de desplazamiento (3 posiciones)
print(f"El abecedario tiene {len(ABECEDARIO)} letras.")

# ----------------------------------------------------

def cifrar_mensaje(mensaje, clave):
    """Cifra el mensaje usando un desplazamiento (Cifrado César)."""
    mensaje_cifrado = ""
    
    # Estandariza el mensaje 
    mensaje_procesado = mensaje.upper()

    # Itera sobre el mensaje
    for letra in mensaje_procesado:
        
        #  If para manejar espacios 
        if letra == ' ':
            mensaje_cifrado += ' '
            continue
            
        #  Encuentra la posición de la letra 
        try:
            indice_original = ABECEDARIO.index(letra)
        except ValueError:
            # caracteres no alfabéticos (ej., números, signos)
            mensaje_cifrado += letra
            continue

        # Aplica la lógica del desplazamiento
        # El módulo asegura que el índice 'vuelva' al inicio si se pasa de 'Z'
        nuevo_indice = (indice_original + clave) % len(ABECEDARIO)

        #  Obtiene la nueva letra y construir el mensaje
        letra_cifrada = ABECEDARIO[nuevo_indice]
        mensaje_cifrado += letra_cifrada
        
    return mensaje_cifrado


# Demostración con mensaje de prueba
print("-" * 60)
# Mensaje original
mensaje_original = "APLICACIONES PARA INGENIERIA CIVIL - FRVM"
print(f"Mensaje original: {mensaje_original}")

# Mensaje cifrado
mensaje_cifrado = cifrar_mensaje(mensaje_original, CLAVE_CESAR)
print(f"Mensaje cifrado:  {mensaje_cifrado}")

# Descifrado: se invierte la clave 
clave_descifrado = -CLAVE_CESAR
mensaje_descifrado = cifrar_mensaje(mensaje_cifrado, clave_descifrado)

print(f"\nClave de descifrado: {clave_descifrado}")
print(f"Mensaje descifrado: {mensaje_descifrado}")

print("-" * 60)

```
